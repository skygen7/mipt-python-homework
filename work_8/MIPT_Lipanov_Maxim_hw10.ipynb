Задание 4
В вашем распоряжении массив A - двумерный numpy.array со абсолютными значениями некоторого набора параметров.
По колонкам записаны отдельные параметры.
Ваша задача - для каждого из параметров произвести стандартизацию.


```python
np.random.seed(777)
A = np.random.randint(1, 180, (100, 15))
```

```python
def std_f(x):
    std = (x - np.mean(x)) / np.std(x)
    return std

std_arr = np.apply_along_axis(std_f, 0, A)
print(std_arr)
```

В университете Вова плохо учил статистику и покупает лотерейные билеты.
Несмотря на это, Вова хорошо знает программирование, поэтому он записал все купленные им билеты в массив numpy.array lottery.
В этом массиве в строках перечислены числа в билете (их всегда одинаковое количество).
Выведите на экран все билеты, в которых отмечено несчастливое для Вовы число 13. Выводить следует по одному билету в одной строке.

```python
np.random.seed(777)
lottery = np.random.choice(90, size=(50, 30))
```

```python
res = np.where(lottery == 13)
print(lottery[np.unique(res[0])])
```
